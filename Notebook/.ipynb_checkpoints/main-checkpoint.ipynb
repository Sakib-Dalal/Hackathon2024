# 4. Feature Engineering

### Selecting Features 
We are going to use correlation matrix to select the feature in the `health_df` dataframe. But before Selecting features the column `Patient_Name` is not important for predicting the target column `Disease`. So we've drop `Patient_Name` column.

```python
health_df.drop("Patient_Name", axis=1, inplace=True)
```

```python
health_df.columns
```

```python
health_df.head()
```

**Correlation Matrix Of Data Before Scalling**

```python
bs_corr = health_df.iloc[:, 1:4].corr(method = "pearson")
```

```python
bs_corr
```

Visualizing with heat map using seaborn

```python
sns.heatmap(bs_corr, annot=True);
```

Hence our target feature is `Disease` which is an object data type, the correlation matrix does not consider the target feature. To prevent this we need to convert our target `Disease` attribute into numberical data type.

Before Scalling our `health_df` dataframe we need to Split the data into X and y where,
* X is our independent features
* y is our dependent features

* X consist `Weight_kg`, `Height_cm`, `Blood_Pressure_mmHg` columns.
* y consist `Disease` column.

```python
X = health_df.drop("Disease", axis=1)
```

```python
X.head()
```

```python
y = health_df["Disease"]
```

```python
y.head()
```

Now we can split our X and y data into training and testing data as `(X_train, y_train)` and `(X_test, y_test)`. We know that after index 751 the data is testing data and before 751 data is training data.

```python
y.count()
```

```python
X_train, X_test, y_train, y_test = X[:750], X[750:], y[:750], y[750:]
```

```python
len(X_train), len(X_test)
```

```python
len(y_train), len(y_test)
```

Now we can combine our train data and test data 

```python
X_train["Disease"] = y_train
train_df = X_train
```
